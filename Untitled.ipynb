# How do ice sheet's filter climate forcing?
To build intuition for this problem we can turn to kinematic moodels of marine outlet glaciers. We'll start by building a timeseries of white noise variability in SMB with a mean of .5m/year with a standard deviation of .1 m/year.


```python
# Simulation parameters
N = 2000 # numer of years in our simulation
Sbar = .5 # 
Sσ = Sbar/5.
Obar = forcing.Ωbar(C,A,θ,simulation.m(n))
Oσ = 0.0
dt = 1.0
```

```python
Ω,smb=simulation.initialize_forcing(Sbar,Sσ,Obar,0.0,N)
```

```python
fig,ax,plots=plot.makefig01(1)
plot.figure01(fig,ax,smb*year,N,dt,label=r'smb.($ma^{-1}$)')
ax.grid(True)
```

```python
b0=100
m0=-2e-3
bedn=3
timescalesm2=[]
H0,L0=simulation.equilibrium(1413,184000,m0,b0,forcing.Ωbar(C,A,θ,simulation.m(bedn)),Sbar/year,C=C,bedn=bedn,alphan=3,n=3)
print(H0)
print(L0)
```

```python
fig1,ax1,ax2,plots=plot.makefig02(1)
H,L,Q,Qg=simulation.simulation(H0,L0,m0,b0,Ω,smb,N)
plot.figure02(fig1,ax1,ax2,H,L/1000,N,dt)
ax1.grid(True)
ax2.grid(True)
```

```python
# Simulation parameters
N = 1000000 # numer of years in our simulation
Sbar = .5 # 
Sσ = Sbar/5.
Obar = forcing.Ωbar(C,A,θ,simulation.m(n))
Oσ = 0.0
dt = 1.0
Ω,smb=simulation.initialize_forcing(Sbar,Sσ,Obar,0.0,N)
```

```python
C_slippery=C/5
C_resistive=C*5
```

```python
b0=100
m0=-2e-3
bedn=3
timescalesm2=[]
H0,L0=simulation.equilibrium(1413,184000,m0,b0,forcing.Ωbar(C,A,θ,simulation.m(bedn)),Sbar/year,C=C,bedn=bedn,alphan=3,n=3)
print(H0)
print(L0)
```

```python
H,L,Q,Qg=simulation.simulation(H0,L0,m0,b0,Ω,smb,N)
fig3,ax3,plots=plot.makefig01(1)
plot.figure03(fig3,ax3,smb*L,N,dt)
plot.figure03(fig3,ax3,Qg,N,dt)
ax1.grid(True)
ax2.grid(True)
```

```python
H,L,Q,Qg=simulation.simulation(H0,L0,m0,b0,Ω,smb,N)
H0,L0=simulation.equilibrium(1413,184000,m0,b0,forcing.Ωbar(C_slippery,A,θ,simulation.m(bedn)),Sbar/year,C=C_slippery,bedn=bedn,alphan=3,n=3)
H_s,L_s,Q_s,Qg_s=simulation.simulation(H0,L0,m0,b0,Ω,smb,N,C=C_slippery)
H0,L0=simulation.equilibrium(1413,184000,m0,b0,forcing.Ωbar(C_resistive,A,θ,simulation.m(bedn)),Sbar/year,C=C_resistive,bedn=bedn,alphan=3,n=3)
H_r,L_r,Q_r,Qg_r=simulation.simulation(H0,L0,m0,b0,Ω,smb,N,C=C_resistive)
fig3,ax3,plots=plot.makefig01(1)
plot.figure03(fig3,ax3,smb*L,N,dt)
plot.figure03(fig3,ax3,Qg,N,dt)
plot.figure03(fig3,ax3,Qg_s,N,dt)
plot.figure03(fig3,ax3,Qg_r,N,dt)
ax1.grid(True)
ax2.grid(True)
```
